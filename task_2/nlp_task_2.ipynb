###### 1.1分词的概念

        词是最小的能够独立活动的有意义的语言成分，英文单词之间以空格作为自然分界符，而汉语是以字为基本的书写单位，词语之间没有明显的区分标记，因此中文词语分析是中文信息处理的进出与关键。常见的方法有：正向最大，逆向最大、双向最大匹配法。
* 1 正向最大：从前往后取词，每次减一个字，直至词典命中或剩下1个单字。
* 2 逆向最大：从后往前取词，每次减一个字，直至词典命中或剩下1个单字。
* 3 双向最大匹配：正向最大与逆向最大两种算法都进行一遍 分词，根据词的<font color="#dd0000">**词的颗粒度越大于好**</font>且<font color="#dd0000">**单字和非字典词越少越好**</font>，进行分词。

# 2.语言模型

###### n-gram语言模型
    n-gram模型也称为n-1阶马尔科夫模型，它有有一个有限历史假设：当前词的出现概率仅仅与前面n-1个此相关： 
\begin{equation} P(s)=P(W_1,W_2,...,W_k)=\prod_{i=1}^kp(W_i|W_i-n+1,...,W_i-1) \end{equation}


###### n-gram模型的参数估计
    模型的参数估计也称为模型的训练，一般采用最大似然估计的方法对模型的参数进行估计：
$$P(W_i|W_i-n+1,\ldots,W_i-1)=\frac{C(W_i-n+1,\ldots,W_i-1,W_i)}{C(W_i-n+1,\ldots,W_i-1)}$$

```python
class NGram(object):

    def __init__(self, n):
        # n is the order of n-gram language model
        self.n = n
        self.unigram = {}
        self.bigram = {}
    # caluclate the ngram of the words
    #
    # @param    words       list{str}
    # @return   none
        # scan a sentence, extract the ngram and update their
    # frequence.
    #
    # @param    sentence    list{str}
    # @return   none
    def scan(self, sentence):
        # file your code here
        for line in sentence:
            self.ngram(line.split())
        #unigram
        if self.n == 1:
            try:
                fip = open("D:/ML_Study/task_2/cnews/uni.txt","w")
            except:
                print >> sys.stderr ,"failed to open uni.txt"
            for i in self.unigram:
                fip.write("%s %d\n" % (i,self.unigram[i]))
        if self.n == 2:
            try:
                fip = open("D:/ML_Study/task_2/cnews/bi.txt","w")
            except:
                print >> sys.stderr ,"failed to open bi.txt"
            for i in self.bigram:
                fip.write("%s %d\n" % (i,self.bigram[i]))
                
    def ngram(self, words):
        # unigram
        if self.n == 1:
            for word in words:
                if word not in self.unigram:
                    self.unigram[word] = 1
                else:
                    self.unigram[word] = self.unigram[word] + 1

        # bigram
        if self.n == 2:
            num = 0
            stri = ''
            for i in words:
                num = num + 1
                if num == 2:
                    stri  = stri + " "
                stri = stri + i
                if num == 2:
                    if stri not in self.bigram:
                        self.bigram[stri] = 1
                    else:
                        self.bigram[stri] = self.bigram[stri] + 1
                    num = 0
                    stri = ''

if __name__=="__main__":
    uni = NGram(1)
    bi = NGram(2)
    uni.scan(file)
    bi.scan(file)
```

###### 文本矩阵化

```python
import jieba
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer 
# 读取停用词
def read_stopword(filename):
    stopword = []
    fp = open(filename, 'r',encoding='utf-8')
    for line in fp.readlines():
        stopword.append(line.replace('\n', ''))
    fp.close()
    return stopword
# 切分数据，并删除停用词
def cut_data(data, stopword):
    words = []
    for content in data['content']:
        word = list(jieba.cut(content))
        for w in list(set(word) & set(stopword)):
            while w in word:
                word.remove(w)
        words.append(' '.join(word))
    data['content'] = words
    return data
# 获取单词列表
def word_list(data):
    all_word = []
    for word in data['content']:
        all_word.extend(word)
    all_word = list(set(all_word))
    return all_word
# 计算文本向量
def text_vec(data):
    count_vec = CountVectorizer(max_features=300, min_df=2)
    count_vec.fit_transform(data['content'])
    fea_vec = count_vec.transform(data['content']).toarray()
    return fea_vec

if __name__ == '__main__':
    data = pd.read_csv('D:/ML_Study/task_2/cnews/cnewstestpart.txt', names=['title', 'content'], sep='\t') 
 
    stopword = read_stopword('D:/ML_Study/task_2/cnews/cnews.vocab.txt')
    data = cut_data(data, stopword)
 
    fea_vec = text_vec(data)
    print(fea_vec)
```
